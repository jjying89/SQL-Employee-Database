```python
#import dependecies
from sqlalchemy import create_engine
import pandas as pd
import matplotlib.pyplot as plt
import os

#get username and password from windows environment variables
username = os.environ.get('db_user')
password = os.environ.get('db_pass')
```

```python
#create engine and make connection
engine = create_engine(f'postgresql://{username}:{password}@localhost:5432/sql-challenge')
connection = engine.connect()
```

```python
#import and check that tables can be read
employees = pd.read_sql("SELECT * FROM employees", connection)
salaries = pd.read_sql("SELECT * FROM salaries", connection)
titles = pd.read_sql("SELECT * FROM titles", connection)
employees.head()
```

```python
#merge employees and salaries
df = pd.merge(employees, salaries, how='inner', on='emp_no')
df.head()
```

```python
#rename column name for merging
titles.columns = ['emp_title', 'title']
titles
```

```python
#merge titles into df
df = pd.merge(df, titles, how='inner', on='emp_title')
df
```

```python
#view salaries
salaries
```

```python
#plot most common salary ranges
histogram = salaries.plot.hist(y='salary', rot=90, legend=False)
histogram.set_xlabel("Salary")
histogram.set_ylabel("Count")
```

```python
#get avg salary by title
avg_salary = df.groupby('title').mean()
avg_salary
```

```python
#plot avg salary by title
bar_plot = avg_salary.plot.bar(y='salary', rot=90, legend=False)
bar_plot.set_xlabel("Title")
bar_plot.set_ylabel("Salary")
```
